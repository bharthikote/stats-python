```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample, kde= True)
    plt.subplot(1,2,2)
    stat.probplot(sample, dist= "norm", plot = pylab)
    plt.show()
```

```python
from matplotlib import *
```

```python
plot_data(data)
```

```python
plot_data(data_copy)
```

```python
plot_data(s)
```

```python
plot_data(p)
```

```python
plot_data(np.log(p))
```
